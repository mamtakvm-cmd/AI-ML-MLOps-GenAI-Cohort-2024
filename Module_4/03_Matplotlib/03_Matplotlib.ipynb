# Bar Chart

```python
import numpy as np
```

```python
company = ['Google','Amazon','MS','Facebook']
revenue = [90, 136, 89, 27]
```

```python
xpos = np.arange(len(company))

xpos
```

```python
plt.bar(xpos,revenue,label = 'Revenue')
plt.ylabel('Revenue(Bln)')
plt.title('Technology Company Stocks')
plt.legend()
```

```python
plt.bar(xpos,revenue,label = 'Revenue')

plt.xticks(xpos,company)

plt.ylabel('Revenue(Bln)')
plt.title('Technology Company Stocks')
plt.legend()
```

```python
profit = [40,2,34,12]
```

```python
plt.bar(xpos,revenue,label = 'Revenue')
plt.bar(xpos,profit,label = 'Profit')
plt.ylabel('Revenue(Bln)')
plt.title('Technology Company Stocks')
plt.legend()
```

```python
plt.bar(xpos - 0.2,revenue,label = 'Revenue')
plt.bar(xpos + 0.2,profit,label = 'Profit')
plt.ylabel('Revenue(Bln)')
plt.title('Technology Company Stocks')
plt.legend()
```

```python
plt.bar(xpos - 0.2,revenue,width=0.4,label = 'Revenue')
plt.bar(xpos + 0.2,profit,width=0.4,label = 'Profit')
plt.ylabel('Revenue(Bln)')
plt.title('Technology Company Stocks')
plt.legend()
```

# Histogram

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
```

```python
plt.hist(blood_sugar, rwidth = 0.8)
```

```python
plt.hist(blood_sugar, rwidth = 0.8, bins= 4)
```

```python
plt.xlabel('Sugar Level')
plt.ylabel('No of Patients')
plt.title('Blood Sugar Levels')

plt.hist(blood_sugar,bins=[80,100],rwidth=0.9,color='b')
plt.hist(blood_sugar,bins=[100,125],rwidth=0.9,color='g')
plt.hist(blood_sugar,bins=[125,150],rwidth=0.9,color='r')
```
